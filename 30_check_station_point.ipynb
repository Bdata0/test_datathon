```python
import pandas as pd
import folium
```

```python
mobile_station_data = pd.read_csv('data/tranformed/data-49158-2020-10-16.csv', 
                                  index_col='Unnamed: 0')
```

```python
mobile_station_data.head()
```

```python
stationare_station_data = pd.read_csv('data/original/data-49133-2020-09-04.csv', 
                    sep=';', 
                    encoding='cp1251', 
                    index_col='ID')
```

```python
stationare_station_data.head()
```

```python
map = folium.Map(location=[55.751244, 37.618423], zoom_start=10, zoom_min=7)
```

```python
for location in mobile_station_data[['Latitude_WGS84', 'Longitude_WGS84']].values:
    folium.Marker(location).add_to(map)
```

```python
for location in stationare_station_data[['Latitude_WGS84', 'Longitude_WGS84']].values:
    folium.Marker(location, icon=folium.Icon(color='red', icon='info-sign')).add_to(map)
```
